```python

mass_radius_powerlaw_model = """

transformed data {    

    m.sun <- 332996.4274 #Msun in Mearth units
    r.sun <- 110.0 # Rsun in Rearth units
    #m.star.const <- 1.37*m.sun
    #
    min.radius.s <- 0.5*r.sun #put priors over this
    max.radius.s <- 2.0*r.sun  #put priors over this
    #
    min.radius.p <- 0.1
    max.radius.p <- 10.0
    #radius.p.exp ~ dunif(0.0,2.0)
    #
    min.mass.s <- 0.8*m.sun #put priors over this
    max.mass.s <- 3.0*m.sun  #put priors over this
    # 
    # Constants (depends on P.ext, P.int, j)
    pi <- 3.14
    #radius.s.true <- r.sun
    #mass.s.true <- m.sun

}

data {}

parameters {}

model {}

    mass.const ~ uniform(0.0,3.0);
    mass.radius.exp ~ uniform(0.0,2.0);
    sigma.mass.radius ~ uniform(0.0,3.0); 
    

    // Population parameters
    for (j in 1:2) {
        e.sigma[j] ~ dunif(0, 1.0)
        // e.sigma[j] <- c(0.001,0.01)
        e.phi[j] <- 1/(e.sigma[j]*e.sigma[j])
        a[j] <- 1;
    }
    //f <- c(0.5,0.5)
    f ~ ddirch(a[])


    for (i in 1:Ndata) {
        C1[i] <- P.int[i]/((pi*(j[i]^(2/3))*((j[i]-1.0)^1/3)*delta[i]))
        C2[i] <- 3.0/(2.0*delta[i])
        C3[i] <- P.ext[i]/(pi*j[i]*delta[i])

    
        // True planet properties
        c[i] ~ dcat(f[]) 
        h.true.int[i] ~ dnorm(0, e.phi[c[i]]) %_% T(-1,1) #Eventually do multivariate truncated normal
        k.true.int[i] ~ dnorm(0, e.phi[c[i]]) %_% T(-sqrt(1-h.true.int[i]*h.true.int[i]),sqrt(1-h.true.int[i]*h.true.int[i]))
        h.true.ext[i] ~ dnorm(0, e.phi[c[i]]) %_% T(-1,1) #Eventually do multivariate truncated normal
        k.true.ext[i] ~ dnorm(0, e.phi[c[i]]) %_% T(-sqrt(1-h.true.ext[i]*h.true.ext[i]),sqrt(1-h.true.ext[i]*h.true.ext[i]))
	
	radius.s.true[i] ~ dunif(min.radius.s, max.radius.s)
        radius.s.obs[i] ~ dnorm(radius.s.true[i], 1.0/(sigma.radius.s[i]*sigma.radius.s[i])) %_% T(0,)
        mass.s.true[i] ~ dunif(min.mass.s,max.mass.s)
        mass.s.obs[i] ~ dnorm(mass.s.true[i],1.0/(sigma.mass.s[i]*sigma.mass.s[i]))

        radius.p.true.int[i] ~ dunif(min.radius.p, max.radius.p) 
        #radius.p.true.int[i] ~ dpar(radius.p.exp,min.radius.p) #%_% T(min.radius.p,max.radius.p)
        radius.ratio.true.int[i] <- radius.p.true.int[i]/radius.s.true[i]
        radius.ratio.obs.int[i] ~ dnorm(radius.ratio.true.int[i], 1.0/(sigma.radius.ratio.obs.int[i]*sigma.radius.ratio.obs.int[i]))
 
        #mass.ref.int[i] <- radius.p.true.int[i]^mass.radius.exp
	mass.ref.int[i] <- mass.const*radius.p.true.int[i]^mass.radius.exp

        mass.p.true.int[i] ~ dnorm(mass.ref.int[i],1.0/(sigma.mass.radius*sigma.mass.radius)) %_% T(0,)
        mass.ratio.true.int[i] <- mass.p.true.int[i]/mass.s.true[i]

        va.true.int[i] <- ((-1.0)*C1[i]*mass.ratio.true.ext[i]*fc[i]) - (C1[i]*C2[i]*mass.ratio.true.ext[i]*((fc[i]*h.true.int[i])+(g[i]*h.true.ext[i])))
        va.obs.int[i] ~ dnorm(va.true.int[i],1.0/(sigma.va.obs.int[i]*sigma.va.obs.int[i]))
        vb.true.int[i] <- C1[i]*C2[i]*mass.ratio.true.ext[i]*((fc[i]*k.true.int[i])+(g[i]*k.true.ext[i]))
        vb.obs.int[i] ~ dnorm(vb.true.int[i],1.0/(sigma.vb.obs.int[i]*sigma.vb.obs.int[i]))
   
        #

        radius.p.true.ext[i] ~ dunif(min.radius.p, max.radius.p) 
        #radius.p.true.ext[i] ~ dpar(radius.p.exp,min.radius.p) #%_% T(min.radius.p, max.radius.p)
        radius.ratio.true.ext[i] <- radius.p.true.ext[i]/radius.s.true[i]
        radius.ratio.obs.ext[i] ~ dnorm(radius.ratio.true.ext[i], 1.0/(sigma.radius.ratio.obs.ext[i]*sigma.radius.ratio.obs.ext[i]))
 
        #mass.ref.ext[i] <- radius.p.true.ext[i]^mass.radius.exp
        mass.ref.ext[i] <- mass.const*radius.p.true.ext[i]^mass.radius.exp

        mass.p.true.ext[i] ~ dnorm(mass.ref.ext[i],1.0/(sigma.mass.radius*sigma.mass.radius)) %_% T(0,)
        mass.ratio.true.ext[i] <- mass.p.true.ext[i]/mass.s.true[i]

        va.true.ext[i] <- ((-1.0)*C3[i]*mass.ratio.true.int[i]*g[i]) + (C2[i]*C3[i]*mass.ratio.true.int[i]*((fc[i]*h.true.int[i])+(g[i]*h.true.ext[i])))
        va.obs.ext[i] ~ dnorm(va.true.ext[i],1.0/(sigma.va.obs.ext[i]*sigma.va.obs.ext[i]))
        vb.true.ext[i] <- (-1.0)*C2[i]*C3[i]*mass.ratio.true.int[i]*((fc[i]*k.true.int[i])+(g[i]*k.true.ext[i]))
        vb.obs.ext[i] ~ dnorm(vb.true.ext[i],1.0/(sigma.vb.obs.ext[i]*sigma.vb.obs.ext[i]))
    }
}

"""

data = {}

#sm = StanModel(model_code=mass_radius_powerlaw_model)
#fit = sm.sampling(data=data, iter=10, chains=2, init=init, n_jobs=-1)

fit = pystan.stan(model_code=mass_radius_powerlaw_model, data=data, iter=1, chains=1, n_jobs=-1);

la = fit.extract(permuted=True)  # return a dictionary of arrays
alpha = la['mass.radius.exp']
beta = la['mass.const']
lnf0 = la['sigma.mass.radius']
print(mass.radius.exp)
print(mass.const)
print(sigma.mass.radius)

a = fit.extract(permuted=False)
print(a)
print(fit)

print(mass.radius.exp)

plt.hist(mass.radius.exp)
fit.plot()


```
